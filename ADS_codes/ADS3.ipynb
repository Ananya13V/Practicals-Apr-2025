```python
# rugplot
import seaborn as sns
import matplotlib.pyplot as plt
data = df
data.head(5)
plt.figure(figsize=(15,5))
sns.rugplot(data=data, x ="Total")
plt.show()
```

```python

#column chart
# plot between 2 attributes
df1 = df.head(10).copy()
df1.plot.bar()
plt.bar(df1['Gender'], df1['Total'])
plt.xlabel("Gender")
plt.ylabel("Total")
plt.show()
```

```python
import plotly.express as px
#Line Plot
df1=df.head(15)
fig = px.line(df1, x="Date", y="Total", color='City')
fig.show()
```

```python
#Bubble Chart
import plotly.express as px

fig = px.scatter(df1, x="Total", y="Tax 5%", size="Quantity", color="City", hover_name="Product line", log_x=True, size_max=60)
fig.show()
```

```python
# Parallel
import plotly.express as px
##Difficult
if not hasattr(pd.DataFrame, "iteritems"):
    pd.DataFrame.iteritems = pd.DataFrame.items
    
df1 =df.sample(n=100)
fig = px.parallel_coordinates(df1, color="Total",
                              dimensions=['Quantity','Unit price','Rating',],
                              color_continuous_scale=px.colors.diverging.Tealrose,
                              color_continuous_midpoint=2)
fig.show()
```

```python
# Creating Andrews curves
df1 = df[['Quantity','Total','Rating']]
df1=df1.sample(n=50)
x = pd.plotting.andrews_curves(df1,'Rating')
 
# plotting the Curve
x.plot()
 
# Display
plt.show()
```

```python
##Difficult

import plotly.express as px
# Monkey patch for pandas >= 2.0 compatibility with plotly
if not hasattr(pd.DataFrame, "iteritems"):
    pd.DataFrame.iteritems = pd.DataFrame.items

# Sample 100 rows but keep all columns
df1 = df.sample(n=100).copy()

# Convert 'Date' to datetime for proper plotting
df1["Date"] = pd.to_datetime(df1["Date"])

# Sort by date for a nice line chart
df1 = df1.sort_values("Date")

# Plot line chart
fig = px.line(df1, x="Date", y="Total", title="Total Sales Over Time (Sampled)")
fig.show()
```
