```python
na_variables = [ var for var in df.columns if df[var].isnull().mean() > 0 ]
#for finding null values in cols
na_variables
```

```python
# mean imputation
df1 = Main_df.copy()

missing_col = ["LoanAmount"]
df1["LoanAmount"].fillna(df1["LoanAmount"].mean(), inplace=True)

df1
```

```python
# median imputation
df2=Main_df.copy()

df2["LoanAmount"].fillna(df2["LoanAmount"].median(), inplace=True)

df2
```

```python
# Mode imputation
df4 = Main_df.copy()
missing_col = ["LoanAmount"]

df4["LoanAmount"].fillna(df4["LoanAmount"].mode()[0], inplace=True)

df4
```

```python
#categorical to numerical 
from sklearn.preprocessing import OrdinalEncoder

data=Main_df.copy()
oe =OrdinalEncoder()
result = oe.fit_transform(data)
print(result)
```

```python
#Random Sample Imputation
df5 = Main_df.copy()

# Get the number of missing values
num_missing = df5["LoanAmount"].isnull().sum()

# Create the random sample from non-null values
random_sample = df5["LoanAmount"].dropna().sample(num_missing, random_state=0)

# Reset index so it can align properly
random_sample.index = df5[df5["LoanAmount"].isnull()].index

# Replace NaNs with random values
df5.loc[df5["LoanAmount"].isnull(), "LoanAmount"] = random_sample

df5
```

```python
# frequent category imputation (for imputation of column with categorical values instead of numeric)
df6=Main_df.copy()
m= df6["Gender"].mode()
m=m.tolist()

frq_imp = df6["Gender"].fillna(m[0])
frq_imp.unique()
```

```python
from sklearn.linear_model import LinearRegression
# Make a copy for imputation
df_reg_imp = Main_df.copy()

# Select only relevant columns
df1 = df_reg_imp[["CoapplicantIncome", "LoanAmount"]]

# Split into training (non-null) and test (null) sets
train_df = df1[df1["LoanAmount"].notnull()]
test_df = df1[df1["LoanAmount"].isnull()]

# Reshape features for sklearn
X_train = train_df["CoapplicantIncome"].values.reshape(-1, 1)
y_train = train_df["LoanAmount"].values

X_test = test_df["CoapplicantIncome"].values.reshape(-1, 1)

# Initialize and fit model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predict missing LoanAmount values
predictions = lr.predict(X_test)

# Assign back to missing rows
missing_indices = df_reg_imp[df_reg_imp["LoanAmount"].isnull()].index
df_reg_imp.loc[missing_indices, "LoanAmount"] = predictions

print("✅ Regression imputation completed.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a mask to distinguish original vs imputed
imputed_mask = (
    df["LoanAmount"].isnull() == False
)  # After assignment, everything is filled

# Plot original + imputed points
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="CoapplicantIncome",
    y="LoanAmount",
    label="After Imputation",
    color="blue",
)

# Overlay the regression line
sns.regplot(
    data=df,
    x="CoapplicantIncome",
    y="LoanAmount",
    scatter=False,
    color="red",
    label="Regression Line",
)

plt.title("LoanAmount vs CoapplicantIncome (with Imputed Values)")
plt.xlabel("Coapplicant Income")
plt.ylabel("Loan Amount")
plt.legend()
plt.show()
```
